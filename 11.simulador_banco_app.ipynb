```python
# Paso 1: Crear el balance inicial y el nombre
import random 
import os 
import csv 
from datetime import datetime

balance = round(random.uniform(500, 1000), 2)
nombre = "Carlos"
print(f"Balance inicial: {balance}, Nombre: {nombre}")
```

```python
# Paso 2: Inicializar listas vacías
historial_balance = []
ingresos = []
salidas = []
```

```python
# Paso 3: Definir la función mostrar_balance
def mostrar_balance(balance):
    print(f"Hola {nombre}! Tu balance actual es {balance}.")
    if balance == 0:
        print("Balance neutro.")
    elif balance < 0:
        print("Balance negativo.")
    else:
        print("Balance positivo.")
```

```python
#Paso 4: Definir la función ingresar_dinero
def ingresar_dinero(balance):
    try:
        ing_banco = float(input("Introduce la cantidad de dinero a ingresar: "))
        if ing_banco <= 0:
            raise ValueError("La cantidad debe ser positiva.")
    except ValueError as e:
        print(f"Error: {e}")
        return balance
    
    ing_info = [ing_banco, "ingreso"]
    ingresos.append(ing_info)
    
    nuevo_balance = balance + ing_banco
    historial_balance.append(nuevo_balance)
    
    mostrar_balance(nuevo_balance)
    return nuevo_balance

```

```python
# Paso 5: Definir la función pagar_servicio
def pagar_servicio(balance):
    try:
        nombre_servicio = input("Introduce el nombre del servicio a pagar: ")
        coste_servicio = float(input("Introduce el coste del servicio: "))
        if coste_servicio <= 0:
            raise ValueError("El coste del servicio debe ser positivo.")
    except ValueError as e:
        print(f"Error: {e}")
        return balance
    
    servicio_info = [coste_servicio, nombre_servicio]
    salidas.append(servicio_info)
    
    nuevo_balance = balance - coste_servicio
    historial_balance.append(nuevo_balance)
    
    mostrar_balance(nuevo_balance)
    return nuevo_balance


```

```python
# Paso 6: Definir la función mostrar_historial_balance
def mostrar_historial_balance(historial_balance):
    for indice_operacion, balance in enumerate(historial_balance):
        print(f"INDICE OPERACION: {indice_operacion}, BALANCE: {balance} €")
    print(f"TOTAL DE OPERACIONES: {len(historial_balance)}, BALANCE ACTUAL: {historial_balance[-1]} €")


```

```python
# Paso 7: Definir la función dicc_balance
def dicc_balance(historial_balance):
    balances = {}
    for indice_operacion, balance in enumerate(historial_balance):
        balances[indice_operacion] = balance
    return balances


```

```python
'''
Paso 8: Función extra
Se define una función para calcular el balance promedio basado en los ingresos y salidas, para analizar patrones de gasto.
'''
def calcular_balance_promedio(ingresos, salidas):
    total_ingresos = sum([ing[0] for ing in ingresos])
    total_salidas = sum([salida[0] for salida in salidas])
    total_transacciones = len(ingresos) + len(salidas)
    
    if total_transacciones == 0:
        return 0
    
    balance_promedio = (total_ingresos - total_salidas) / total_transacciones
    return round(balance_promedio, 2)

```
